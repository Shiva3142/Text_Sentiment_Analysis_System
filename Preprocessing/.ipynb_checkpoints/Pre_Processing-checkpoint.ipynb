# Morphological Processing

```python
Regular_expression_definition_for_html_tags=re.compile('<.*?>')
Regular_expression_definition_for_digits=re.compile('\d+\s|\s\d+|\s\d+\s')
Regular_expression_definition_for_links=re.compile('http://\S+|https://\S+')
word_lemitizer=WordNetLemmatizer()
english_stop_words=stopwords.words('english')
```

```python
def lemmetize_comparer(text):
    text=text.lower().split()
    text=[word for word in text if word not in english_stop_words]
    text_=[word_lemitizer.lemmatize(word) for word in text]
    for i in range(len(text_)):
        print(text[i]," ->",text_[i])
```

```python
def preprocessing_of_sentence(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)
    text=Regular_expression_definition_for_html_tags.sub(r" ",text)
    text=Regular_expression_definition_for_digits.sub(r" ",text)
    text=Regular_expression_definition_for_links.sub(r" ",text)
    punctuations = [".","=","_","<",">",",","!","?","'",'"',":",";","*","-","/","+","%","$","#","@","(",")","[","]","{","}",'\n']
    for i in punctuations:
        text = text.replace(i," ")
    text=text.lower().split()
    text=[word_lemitizer.lemmatize(word) for word in text]
#     text=[word for word in text if len(word)>1]
    return text
```

```python
Preprocessed_training_textual_data=[" ".join(preprocessing_of_sentence(text)) for text in Training_Data[text_column]]
```

# Data Reduction

```python
selected_class_list=[
"admiration",
"anger",
"annoyance",
"approval",
"confusion",
"curiosity",
"desire",
"disappointment",
"disapproval",
"disgust",
"embarrassment",
"excitement",
"fear",
"gratitude",
"joy",
"love",
"nervousness",
"optimism",
"pride",
"realization",
"relief",
"remorse",
"sadness",
"surprise",
]
```

```python
def getSelectedClassesforTrainingData(Data):
    texts=[]
    labels=[]
    for row in range(len(Data)):
        if Data[label_column][row] in selected_class_list:
            texts.append(Data[text_column][row])
            labels.append(Data[label_column][row])
    Data=pandas.DataFrame()
    Data[text_column]=texts
    Data[label_column]=labels
    return Data
```

```python
Preprocessed_Training_Data=getSelectedClassesforTrainingData(Preprocessed_Training_Data)
```

```python
Preprocessed_testing_textual_data=[" ".join(preprocessing_of_sentence(text)) for text in Testing_Data[text_column]]
Preprocessed_Testing_Data=Testing_Data
Preprocessed_Testing_Data[text_column]=Preprocessed_testing_textual_data
```

```python
def getSelectedClassesforTrainingData(Data):
    texts=[]
    labels=[]
    for row in range(len(Data)):
        try:
            if class_list[int(Data[label_column][row])] in selected_class_list:
                texts.append(Data[text_column][row])
                labels.append(class_list[int(Data[label_column][row])])
        except:
            pass
    Data=pandas.DataFrame()
    Data[text_column]=texts
    Data[label_column]=labels
    return Data
```

```python
Preprocessed_Testing_Data=getSelectedClassesforTrainingData(Preprocessed_Testing_Data)
```
